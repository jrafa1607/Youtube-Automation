```python
#Install Tools and Import Libraries
from google.colab import files
import subprocess
import os

%pip install yt-dlp
```

```python
#Function - List all URLs from Youtube
urls = []

while True:
    item = input("Youtube URL: (or 'N' to Stop): ")
    if item.upper() == "N":
        break
    urls.append(item)

print("\nTotal of Musics: " + str(len(urls)))
```

```python
#Function - Music Download for Colab Enviroment
cmd = "yt-dlp -x --audio-format mp3 "

for url in urls:
   run = cmd + url
   subprocess.run(run, shell=True)
urls.clear()
```

```python
#Function - Mp3 Download for Desktop
directory = '/content/'

files_in_directory = os.listdir(directory)
mp3_files = [f for f in files_in_directory if f.endswith('.mp3')]

for mp3_file in mp3_files:
    file_path = os.path.join(directory, mp3_file)
    files.download(file_path)
```

```python
#Function - Delete Musics After Download
for mp3_file in mp3_files:
    file_path = os.path.join(directory, mp3_file)
    os.remove(file_path)

print("All the .Mp3 files are deleted!")
```
